### Checking duplicate columns

```python
# first checking identical columns
# learned from
# https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 

def getDuplicateColumns(EDA_df):

    duplicateColumnNames = set()
    for x in range(EDA_df.shape[1]):
        col = EDA_df.iloc[:, x]
        for y in range(x + 1, EDA_df.shape[1]):
            otherCol = EDA_df.iloc[:, y]
            if col.equals(otherCol):
                duplicateColumnNames.add(EDA_df.columns.values[y])
 
    return list(duplicateColumnNames)


duplicateColumnNames = getDuplicateColumns(EDA_df)
 
print('Duplicate Columns are as follows')
for col in duplicateColumnNames:
    print('Column name : ', col)
```

## 5.3) Checking for missing values
For the purpose of this analysis we have decided to conduct two separate methods when working with missing values. For the purpose of variables like **'construction_year', 'population', 'longitude', 'gps_height', 'amount_tsh'** with the median, since they have a wide range of values. Whilist **'permit'**, since it has two possible values, True and False, we have decided to fill with the mode, in this case it will be True because we had 388,852 variables that were true and 17,492.

- The following columns have over 70% missing: amount_tsh, num_private
- The following columns have a significant proportion missing: gps_height, num_private, population, construnction_year
- Missing values for construction_year are filled with â€œ0â€ which heavily skews the data towards an earlier year


#### 5.3.1) Fill missing values with median


```python
for x in ['construction_year', 'population', 'longitude', 'gps_height', 'amount_tsh']:
    # Calculate median of data excluding zeros
    med = WaterDF[WaterDF[x]!=0][x].median()
    
    # Replace zero value with median
    WaterDF[x] = WaterDF[x].replace(0, med)
```

#### 5.3.2)  Fill missing values with mode

```python
for x in ['permit']:
    # Calculate mode where not NA
    mde = WaterDF[~WaterDF[x].isna()][x].mode().loc[0]
    
    # Replace NA with mode
    WaterDF[x] = WaterDF[x].fillna(mde)
```

## 6.1) Training Models

#### Importing metrics

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

#### Collection of all trained models and results

```python
models = {}
```

```python
def run_model(ID, clf, X_train, X_test, y_train, y_test):  
    # Train model
    clf.fit(X_train, y_train)
    
    # Make predictions
    train_predictions = clf.predict(X_train)
    test_predictions = clf.predict(X_test)
    
    # Score predictions
    train_score = accuracy_score(y_train, train_predictions)
    test_score = accuracy_score(y_test, test_predictions)
    
    confusion_mtx = confusion_matrix(y_test, test_predictions)
    
    # Save model and results to dictionary
    models[ID] = {"clf": clf,
                  "test_score": test_score, 
                  "train_score": train_score,
                  "confusion_mtx": confusion_mtx,
                  "test_predictions": test_predictions
                 }
    print("Test Score: {}".format(test_score))
    print("Train Score: {}".format(train_score))
    print("Confusion Matrix:\n", confusion_mtx)
```

### 6.1.1) Logistic Regression

This modelling tehcnique is used to analyse datasets in which there are one or several  independent variables that can help us determine an outcome. The outcome is usually measured with a dichotomous variable, in other words binary, in which there are only two possible outcomes possible.

```python
from sklearn.linear_model import LogisticRegression
mdl = LogisticRegression(C=100, multi_class='auto', random_state=0)
run_model('LogisticRegression', mdl, X_learn, X_valid, y_learn, y_valid)
```

### 6.1.3) Random Forest

This modelling  technique is a classification algorithm that consists of multiple decision trees. This model uses "bagging" and feature randomness when creating an individual tree to try and create an uncorrelated forestst of trees whose prediction is more accurate than individual trees. Random Forest algorithms increase the predictive power of algorithms and also helps prevent overfitting.

```python
from sklearn.ensemble import RandomForestClassifier
```

Tuning for **max depth:**

```python
from sklearn.ensemble import RandomForestClassifier

for d in range(10, 26, 5):
    print(d)
    mdl = RandomForestClassifier(max_depth=d, n_estimators=100, random_state=0)
    run_model('RandomForestClassifier_maxDepth{}'.format(d), mdl, X_learn, X_valid, y_learn, y_valid)
```

Tuning for **max_features:**


```python
for f in range(1, 13, 2):
    print(f)
    mdl = RandomForestClassifier(max_depth=20, max_features=f, n_estimators=100, random_state=0)
    run_model('RandomForestClassifier_maxDepth20_maxFeatures{}'.format(f), mdl, X_learn, X_valid, y_learn, y_valid) 
```

Maximum score through decision trees reached via maximizing features using 25 splits.
